```python
#The Jacobi method takes more iterations to converge than the Gauss Seidel does. 
# For both methods, the smaller the tolerance, the more iterations needed to converge. 
import numpy as np
import matplotlib.pyplot as plt

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    errors = []
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) 
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        errs =  np.linalg.norm(x_new - x, ord=np.inf)
        errors.append(errs)
        if errs < tol:
            return x_new, k + 1, errors   #chatgpt helped me trouble shoot why my code was not working, it told me to add the k+1 because python starts on 0
        x = x_new
    return x, max_iterations, errors

A = np.array([[3, -0.1, -0.2],
[0.1, 7, -0.3],
[0.3, -0.2, 10]])
b = np.array([7.85, -19.3, 71.4])

x0 = np.zeros_like(b)
max_iterations = 100

tol1 = 1e-3
solution1, iterations1, err1 = gauss_seidel(A, b, x0, tol1, max_iterations)

tol2 = 1e-6
solution2, iterations2, err2 = gauss_seidel(A, b, x0, tol2, max_iterations)


tol3 = 1e-9
solution3, iterations3, err3 = gauss_seidel(A, b, x0, tol3, max_iterations)

print(f"Solution for 1e-3: {solution1}")
print(f"Iterations for 1e-3: {iterations1}")
print(f"Solution for 1e-6: {solution2}")
print(f"Iterations for 1e-6: {iterations2}")
print(f"Solution for 1e-9: {solution3}")
print(f"Iterations for 1e-9: {iterations3}")

x = np.linspace(0, iterations1 + 1, len(err1))
plt.plot (x, err1)
x = np.linspace(0, iterations2 + 1, len(err2))
plt.plot(x, err2)
x = np.linspace(0, iterations3 + 1, len(err3))
plt.plot(x,err3)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    jerrors = []
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        errj =  np.linalg.norm(x_new - x, ord=np.inf) 
        jerrors.append(errj)
        if errj < tol:
            return x_new, k +1, jerrors
        x = x_new
    return x, max_iterations, jerrors

A = np.array([[3, -0.1, -0.2],
[0.1, 7, -0.3],
[0.3, -0.2, 10]])
b = np.array([7.85, -19.3, 71.4])

x0 = np.zeros_like(b)
max_iterations = 100
tol1 = 1e-3
solution1j, iterations1j, err_j1 = jacobi(A, b, x0, tol1, max_iterations)
tol2 = 1e-6
solution2j, iterations2j, err_j2 = jacobi(A, b, x0, tol2, max_iterations)
tol3 = 1e-9
solution3j, iterations3j, err_j3 = jacobi(A, b, x0, tol3, max_iterations)

print(f"Solution for 1e-3: {solution1j}")
print(f"Iterations for 1e-3: {iterations1j}")
print(f"Solution for 1e-6: {solution2j}")
print(f"Iterations for 1e-6: {iterations2j}")
print(f"Solution for 1e-9: {solution3j}")
print(f"Iterations for 1e-9: {iterations3j}")

x = np.linspace(0, iterations1j + 1, len(err_j1))
plt.plot (x, err_j1)
x = np.linspace(0, iterations2j + 1, len(err_j2))
plt.plot(x, err_j2)
x = np.linspace(0, iterations3j + 1, len(err_j3))
plt.plot(x,err_j3)
plt.show()
```

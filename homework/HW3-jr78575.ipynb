```python
import numpy as np
import matplotlib.pyplot as plt

def fact(num):
    sum = 1
    while (num > 1):
        sum *= num
        num -= 1
    return sum

n = 0
avg = 0
aprox = []
x = np.linspace(0, 10,101)

while (n < 4):
    avg += (np.power(-1,n))*(np.power(x,(2*n)+1))/(fact((2*n)+1))
    aprox.append(avg.copy()) #used chat gpt to figure out why the avg wasn't saving in the array which is where I got the .copy from
    n += 1
err1 = abs(aprox[0] - np.sin(x))
err2 = abs(aprox[1] - np.sin(x))
err3 = abs(aprox[2] - np.sin(x))
err4 = abs(aprox[3] - np.sin(x))

plt.plot(x, err1, label = 'One Term')
plt.plot(x, err2, label = 'Two Terms')
plt.plot(x, err3, label = 'Three Terms')
plt.plot(x, err4, label = 'Four Terms')
plt. legend()
plt.show()

#The margin of error decreases as the number of terms increase. However, as x decreases, less terms are more efficient. When x increases, more terms are more efficient.
```

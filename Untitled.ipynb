```python
import numpy as np
import random as rng
import matplotlib.pyplot as plt

def estamatePi (sample):
    count = 0
    trial = 0

    while(trial < sample):
        x = rng.uniform(-1,1)
        y = rng.uniform(-1,1)
        trial += 1
    
        if (x**2 + y**2 < 1):
            count += 1
        else:
            count += 0
            
    est = (4*count) / sample
    return est;

samples = []
errors = []
relErrs = []

for i in range(10,1001,20):
    sample = i
    piEST = estamatePi(sample)
    error = abs(piEST - np.pi) 
    relError = error / np.pi
    
    samples.append(i)
    errors.append(error)
    relErrs.append(relError)
    
plt.plot(samples, errors, label = "Absolute Error")
plt.plot(samples, relErrs, label = "Realative Error")

plt.show
```

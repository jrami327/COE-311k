```python
import matplotlib.pyplot as plt
import numpy as np

a = 0
b = 2*np.pi
N = 100

x = np.linspace(a, b, N)
h = (b - a) / (N - 1)
f = np.sin

df_dx_cd = (f(x + h) - f(x - h)) / (2 * h) #central difference
df_dx_fd = (f(x + h) - f(x)) / h #forward difference
df_dx_bd = (f(x) - f(x-h)) / h #backward difference
df_dx_true = np.cos(x) #true derivative

fig, ax = plt.subplots(2, 2, constrained_layout=True)
ax[0, 0].plot(x, df_dx_true)
ax[0, 0].set_title("True derivative (cos(x))")
ax[0, 1].plot(x, df_dx_fd)
ax[0, 1].set_title("Forward Difference")
ax[1, 0].plot(x, df_dx_bd)
ax[1, 0].set_title("Back Difference")
ax[1, 1].plot(x, df_dx_cd)
ax[1, 1].set_title("Central Difference")
plt.show()
```

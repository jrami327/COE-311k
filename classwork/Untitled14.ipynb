```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
h = [0.1, 0.5, 1, 1.5]  # step size
x_start, x_end = 0, 2
current_x, current_y = None, None

def f(x, y):      # Function representing dy/dx = -2y
    return -2 * y

plt.figure()

for i in range(len(h)):
    hx = h[i]
    num_steps = int((x_end - x_start) / hx)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1  # y(0) = 1     # Initial condition

    for k in range(num_steps):    # Implement Euler's method
        y_values[k + 1] = y_values[k] + hx * f(x_values[k], y_values[k])
    
    y_exact = y_values [0] * np.exp(-2 * x_values)   
    
    local_err = np.abs(y_exact - y_values)
    global_err = np.max(local_err)
    print(f'Step Size: {hx:.2f}       Max Global Error: {global_err:.5f}')
    plt.plot(x_values, local_err, label = f'h = {hx:.2f}')
    
plt.xlabel('x')
plt.ylabel('Local Error')
plt.title("Euler Method Error vs Exact Solution")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

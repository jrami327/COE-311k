```python
import numpy as np
import matplotlib.pyplot as plt
import time

def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

matrix_size = []
size = 3
for i in range(8):
    matrix_size.append(size)
    size *= 2

naive_time = []
np_time = []
rng = np.random.default_rng()

for j in matrix_size:
    a = rng.random((j, j))
    b = rng.random((j, j))
    
    start = time.time()
    naive_matrix_multiplication(a.tolist(), b.tolist())
    naive_time.append(time.time() - start) #chat gpt helped me calculate the time; I read the time and save it as start then read the time again and subtract by the start to get the 
    
    start = time.time()
    np.dot(a, b)
    np_time.append(time.time() - start)
    
x = range(1, 9)
plt.plot(x, naive_time, label = 'Naive Time')
plt.plot(x, np_time, label = 'Numpy Time')
plt.show()

#The naive time takes longer the larger the size of the matrix it is multiplying, whereas the numpy method is more or less maintaining the same time.
```

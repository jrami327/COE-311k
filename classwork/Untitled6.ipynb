```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")

plt.plot(X, m*X +c)
plt.scatter(X,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
m_true = 3
c_true = 5

rng = np.random.default_rng()
X = np.arange(0, 1.01, 0.01)
b = rng.random(len(X))
sign = rng.choice([-1,1], len(X)) #chatgpt helped me figure out how to code the random number, specifically how to code the different sign while continuing the range between 0-1
Y = m_true*X + c_true + b * sign

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
plt.plot(X, m*X + c)
plt.scatter(X,Y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 50)

plt.subplot(6,2,1)
plt.plot(x, np.sin(x), 'r-')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,2,3)
plt.hist(x,50)
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,4)
plt.scatter(x,np.tan(x))

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

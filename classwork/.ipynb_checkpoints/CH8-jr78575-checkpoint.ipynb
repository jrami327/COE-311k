```python
import numpy as np

def matrix_dom(a):
    row = a.shape[0]
    col = a.shape[1]
    add = 0
    for i in range(row):
        for j in range(col):
            if i != j:
                add += abs(a[i,j])
        if abs(a[i,i]) < add:
            return print('False')
        add = 0
    return print('True')

matrix = np.array([[7, 3, 1], [3, 6, 2], [3, 4, 9]])   

print(matrix)
print('Is the matrix diagonally dominant?:')
matrix_dom(matrix)
```

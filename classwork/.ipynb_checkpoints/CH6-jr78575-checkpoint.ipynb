```python
import numpy as np
import time

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

start_time = time.perf_counter()

A = np.array([[1, 2, 3, 0, 0],
[2, 1, 2, 3, 0],
[3, 2, 1, 2, 3],
[0, 3, 2, 1, 2], [0, 0, 3, 2, 1]])
b = np.array([14, 22, 33, 26, 22])

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)

for i in range(1000000):
    pass
end_time = time.perf_counter()
elapsed_time = end_time - start_time

print(f"Elapsed time: {elapsed_time:.4f} seconds")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")

# for 6a it took 2 iterations and the time was 0.0366 seconds; the solution was 1 2 2 3 4
# for 6b it diverged and was 100 iterations; the time was 0.0474
```

```python
import numpy as np
import time

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
# Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

start_time = time.perf_counter()

A = np.array([[1, 2, 3, 0, 0],
[2, 1, 2, 3, 0],
[3, 2, 1, 2, 3],
[0, 3, 2, 1, 2], [0, 0, 3, 2, 1]])
b = np.array([14, 22, 33, 26, 22])

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = jacobi(A, b, x0, tol, max_iterations)

for i in range(1000000):
    pass
end_time = time.perf_counter()
elapsed_time = end_time - start_time

print(f"Elapsed time: {elapsed_time:.4f} seconds")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")

# for 6a it took 1 iteration and the time was 0.0380 seconds; the solution was 1 2 2 3 4
# for 6b it diverged and was 100 iterations; the time was 0.0473
```

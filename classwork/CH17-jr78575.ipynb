```python
import numpy as np
import matplotlib.pyplot as plt
# Define the parameters
k = 1.0        # Constant in the equation (e.g., spring constant)
h = [.01, .02, .03, .05, .1, .2, .3, .5]      # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx
plt.figure(figsize=(10, 5))

# Create arrays to store x, y, and v values
for j in range(len(h)):
    x_values = np.arange(x_start, x_end + h[j], h[j])
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

# Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    
        # Update equations
        y_values[i] = y_n + h[j] * v_n
        v_values[i] = v_n + h[j] * (-k * y_n)
    
    plt.plot(x_values, y_values, label=f'Position y(x); h = {h[j]}')
    plt.plot(x_values, v_values, label=f'Velocity v(x); h = {h[j]}', linestyle='--')
# Plot the results
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
#It starts to become unstable around h = 0.1 we can see this because the oscillation's maxima and minima are growing
#Yes it is becoming unstable as the graph continues and we can see as evident as the statement above. 
#As K increases the frequency increases. 
#No, I don't think it could solve with a larger h. This graph shows that Euler is not very good at solving systems like this and becomes unstable as h increases when trying to solve them. 
```

```python
import numpy as np
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
        n = len(b)
        x = x0.copy()
        for k in range(max_iterations):
            x_new = x.copy()
            for i in range(n):
                s1 = sum(A[i][j] * x_new[j] for j in range(i)) 
                s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) 
                x_new[i] = (b[i] - s1 - s2) / A[i][i]
            if np.linalg.norm(x_new - x, ord=np.inf) < tol:
                return x_new, k
            x = x_new
        return x, max_iterations
    
A = np.array([[1, 2, 3, 0, 0],
            [2, 1, 2, 3, 0],
            [3, 2, 1, 2, 3],
            [0, 3, 2, 1, 2],
             [0, 0, 3, 2, 1]])
b = np.array([14, 22, 33, 26, 22])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
Jsolution, Jiterations = jacobi(A, b, x0, tol, max_iterations)
Gsolution, Giterations = gauss_seidel(A, b, x0, tol, max_iterations)
print(f"Jacobi Solution: {Jsolution}")
print(f"Jacobi Iterations: {Jiterations}")
print(f"Gauss Seidel Solution: {Gsolution}")
print(f"Gauss Seidel Iterations: {Giterations}")

#for challenge 6a- both converge
#Jacobi Solution: [1 2 2 3 4]
#Jacobi Iterations: 1
#Gauss Seidel Solution: [1 2 2 3 4]
#Gauss Seidel Iterations: 2

#for challenge 6b- both diverge
#Jacobi Solution: [-5976176488926883840 -4324294422220818432   125198747742650400 -4324294422220818432 -5976176488926883840]
#Jacobi Iterations: 100
#Gauss Seidel Solution: [-4667072227903264768 -5861557759596381184  5941495520527654912 32542205762157592   557173100602271744]
#Gauss Seidel Iterations: 100
```

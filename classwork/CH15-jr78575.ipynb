```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
h = np.arange(0.1, 1.01, 0.05)  # step size
x_start, x_end = 0, 2
current_x, current_y = None, None

def f(x, y):      # Function representing dy/dx = -2y
    return -2 * y

plt.figure()

for i in range(len(h)):
    hx = h[i]
    num_steps = int((x_end - x_start) / hx)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1  # y(0) = 1     # Initial condition

    for k in range(num_steps):    # Implement Euler's method
        y_values[k + 1] = y_values[k] + hx * f(x_values[k], y_values[k])

    if i == 0:
        current_x = x_values #chat gpt helped me figure out how to impliment the first iteration into the loop without crashing it
        current_y = y_values
        plt.plot(current_x, current_y, label = 'First Run')
        continue 
    
    interp_y = np.interp(current_x, x_values, y_values)
    err = np.abs(interp_y - current_y)
    tot_err = np.sum(err)
    print(f'Step Size: {hx:.2f}       Total Error: {tot_err:.5f}')
    plt.plot(x_values, y_values, label = f'h = {hx:.2f}')
    
plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

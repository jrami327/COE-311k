```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
# Define the ODE
def f(t, y):
    return -(1000 * y) + 3000 -(2000 * np.exp(-t))
def exact(t):
    return 3 - (0.998* np.exp(-1000*t)) - 2.002 * np.exp(-t)

h = [.1, .01, .001]      # Step size

def forward_euler(f, y0, t0, tf, h):
    t_values = np.arange(t0, tf + h, h)
    y_values = np.zeros_like(t_values)
    y_values[0] = y0
    for i in range(1, len(t_values)):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1],
        y_values[i-1])
    return t_values, y_values
# Solve using Forward Euler

for i in h: 
    t_values_fe, y_values_fe = forward_euler(f, 0, 0, 30, i)
    plt.plot(t_values_fe, y_values_fe, label=f'Forward Euler h={i}')

t_exact = np.linspace(0, 30, 1000)
exact_values = exact(t_exact)
plt.plot(t_exact, exact_values, label = 'Analytical Solution')
plt.legend()
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Comparison: Forward vs Analytical Solution')
plt.grid(True)
plt.show()
```
